```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

## Minimum değerler 0 biz onları nulla çeviricez ve doldurucaz 
#### We can notice that have many zeros values. We'll convert to null values and impute it.

```python
data.describe().T
```

```python
data[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = data[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.NaN)
```

```python
data.isna().sum()
```

```python
print(data['SkinThickness'].median())
```

## impute işlemi yapacağım median ve meanlerle

```python
data['Glucose'].fillna(data['Glucose'].mean(),inplace=True)
data['BloodPressure'].fillna(data['BloodPressure'].mean(),inplace=True)
data['SkinThickness'].fillna(data['SkinThickness'].median(),inplace=True)
data['Insulin'].fillna(data['Insulin'].mean(),inplace=True)
data['BMI'].fillna(data['BMI'].mean(),inplace=True)
```

```python
data.describe().T
```

```python
X=data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.2,stratify=y)

print(X_train.shape)
print(X_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
from sklearn.metrics import mean_absolute_error,confusion_matrix,accuracy_score
```

```python
from xgboost import XGBClassifier
```

```python
model=XGBClassifier(objective='multi:softmax',num_class=2,learning_rate=0.06)
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_valid)
```

```python
print(mean_absolute_error(y_valid,predictions))
```

```python
accuracy_score(y_valid,predictions)
```

```python
score=cross_val_score(model,X,y,cv=5)
print(score.mean())
```

```python
print(confusion_matrix(y_valid,predictions))
```

```python
sns.distplot(y_valid,bins=2)
plt.show()
```

## Outlier Detection


```python
for column in data[1:-1]:
    sns.scatterplot(x=column,y='BMI',hue='Outcome',data=data)
    plt.show()
```

```python
data['Pregnancies']
```

```python
outliers_pregnancies=data[data['Pregnancies']>data['Pregnancies'].mean()+3*data['Pregnancies'].std()].index
```

```python
data.drop(index=outliers_pregnancies,inplace=True)
```

```python
outliers_pregnancs=data[data['Pregnancies']>8].index
```

```python
data.drop(index=outliers_pregnancs,inplace=True)
```

```python
data['Pregnancies'].max()
```

```python
outlier_blood=data[data['BloodPressure']>140].index 
outlier_blood
```

```python
data.drop(index=outlier_blood,inplace=True)
```

```python
data['BMI'].max()
```

```python
outliers_bmi=data[data['BMI']>45].index
```

```python
data.drop(outliers_bmi,inplace=True)
```

```python
outliers=data[data['Insulin']>429].index
```

```python
data.drop(outliers,inplace=True)
```

```python
data['Insulin'].max()
```

```python
for column in data[1:-1]:
    sns.scatterplot(x=column,y='BMI',hue='Outcome',data=data)
    plt.show()
```

```python

X=data[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','Age']]
y=data['Outcome']
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid=train_test_split(X,y,train_size=0.8)
```

```python
from xgboost import XGBClassifier
```

```python
model=XGBClassifier(objective='multi:softmax',num_class=3)

model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_valid)
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error,confusion_matrix,accuracy_score
```

```python
print(mean_absolute_error(y_valid,predictions))
```

```python
score=cross_val_score(model,X,y,cv=5)
print(score)
```

```python
score.mean()
```

```python
accuracy_score(y_valid,predictions)
```

```python
print(confusion_matrix(y_valid,predictions))
```
